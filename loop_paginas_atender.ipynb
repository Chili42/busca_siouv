```python
#verificar instalação do selenium
#utilizar o comando !pip install selenium -i http://binario.caixa:8081/repository/pypi-repo/simple --trusted-host binario.caixa --user
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
import getpass
from bs4 import BeautifulSoup
from datetime import datetime

from dotenv import load_dotenv
from dotenv import find_dotenv
load_dotenv(find_dotenv())
```

```python
def login_atender(matricula, senha):
    browser = webdriver.Firefox(executable_path='geckodriver.exe')

    # abrir a página do Atender.caixa
    url = 'https://atender.caixa/siouv/jsp/login/login.jsp'

    browser.get(url)

    form_usuario = browser.find_element_by_name('j_username')
    form_usuario.send_keys(matricula)

    form_senha = browser.find_element_by_name('j_password')
    form_senha.send_keys(senha)

    botao_login = browser.find_element_by_xpath('//form/table/tbody/tr/td/table[1]/tbody/tr[1]/td/div/table/tbody/tr[4]/td/table/tbody/tr/td[2]/a')
    botao_login.click()

    return browser
```

```python
# main
matricula = input('Digite sua matrícula: ')
senha = getpass.getpass('Digite sua senha: ')

browser = login_atender(matricula, senha)
```

```python
time.sleep(5)
browser.find_element_by_link_text('Consulta Ocorrências para Tratamento').click()
time.sleep(5)
```

```python
## url paginação


lista_links = []
url_paginacao= 'https://atender.caixa/siouv/jsp/login/ConsultarOcorrencia.do?nuProtSac=&ufRar=-1&perfilUsuario=gestor&anoRar=&sequencialPreOcorrencia=&cnpj=&sequencialTipoOcorrenciaInterna=1&situacaoClassificacao=&sequencialOcorrencia=&unidadeRemetente=2&periodoFinal=&ordenacaoOuvidoria=1&cpf=&tipoOrdenacaoSac=ASC&sequencialTipoConsulta=1&tipoOcorrenciaClassificacao=&tipoOrdenacaoInterna=ASC&escopo=I&assuntoClassificacao=&ordenacaoSac=1&numeroRar=&coUnidadePesquisa=&itemClassificacao=&tipoOrdenacaoOuvidoria=ASC&gravaHistorico=&nis=&periodoInicial=&guardaTextoUnidadeDestino=&method=consultarOcorrencia&usuarioSolicitante=&motivoClassificacao=&ordenacaoInterna=1&page={}&'
paginas = range(1,9)


for pagina in paginas:
    url = url_paginacao.format(pagina)
    
    
    browser.get(url)
    
    soup = BeautifulSoup(browser.page_source, 'lxml')
    
    tabelas = soup.find_all("table",{"class":"TabelaPadrao"})
    
    # print(tabelas)
    
    for tabela in tabelas:
        
        links = tabela.find_all('a')
        if len(links)== 0:
            continue
        
        lista_links += links
    


```
